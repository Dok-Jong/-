#**1. 고대문자 해석(쿠로넷)**
* 쿠주시지(くずしじ)는 일본의 필기체 글자를 말한다. 1000년 동안 써 온 문자인데, 1900년대 들어서 교과서가 표준화되고 문자를 쓰는 방법이 정립되면서 서서히 필기체가 사라지기 시작했다. 문제는 쿠주시지로 쓰인 책이 산더미처럼 쌓여 있는데 이를 읽을 수 있는 사람이 거의 없다는 것이다. 현재 쿠주시지를 읽을 수 있는 사람은 일본 인구의 0.1% 수준이고 그나마도 계속해서 줄고 있다고 한다.

* 현재 일본에서는 이 쿠주시지로 쓰인 고문서들을 디지털화 하는 프로젝트가 진행되고 있으며, 수천억개의 이미지가 쌓여있다. 전문가들이 이 데이터를 바탕으로 컴퓨터에게 읽을 수 있도록 머신러닝 모델을 만들기 시작했고, 이를 ‘쿠로넷’이라고 이름 붙였다. 일반적인 필기체를 인식하는 OCR형태의 모델이며, 글자를 읽고 해석과 검색까지 할 수 있는 것이 목표이다.

 * 현재 2초에 한 페이지를 읽을 수 있으며, 책 한권을 해석하는데 한 시간 정도 걸린다. 하지만 아직 많은 책을 읽어 내지는 못하고 있다. 정확도는 85%이고 읽을 수 있는 책도 16권 밖에 되지 않는다. 계속해서 모델을 개선해서 많은 책에 적용할 수 있도록 할 계획이다.

> 꾸준한 연구를 통하여, 밝혀낸 단서를 머신러닝을 통하여 아직 밝혀내지 않은 역사의 실체를 밝혀내고, 더 나아가 미스터리에 쌓여있는 외계문자까지 해독이 가능하게 되어 그들의 문명을 이해하며 소통하게 된다면 정말 이보다 흥미로운 것은 없을 것 같습니다.

#**4. 자율주행 자동차(ADAS와 자율주행)**
* 자율주행 자동차에 대해서 언급하려면 우선, 'ADAS(Adanced Driver Assistance Systems)‘에 대해서 알 필요가 있다. 
 ADAS센서는 운전 중 발생할 수 있는 수많은 상황 중 일부를 차량 스스로 인지하고 상황을 판단하여, 기계장치의 제어에 개입을 하여 운전자를 지원해주는 시스템이다.
* ADAS의 가장 큰 장점은 객체와 차량, 교통 환경을 모두 감지하여 이에 대응할 상황을 예측 할 수 있는 것이다. 그 기능에는 긴급제동 시스템(AEB), 주행 조향보조 시스템(LKAS), 어드밴스드 스마트 크루즈 컨트롤(ASCC), 후측방 충돌 회피 지원 시스템(ABSD), 어라운드 뷰 모니터링 시스템(AVM)을 포함합니다. 

 * > ADAS 와 자율주행, 같을까 다를까? 
자율주행 기술에는 5단계가 있다고 한다. 이 중에서 ADAS는 어디에 해당될까? 누군가는 2~3단계, 또 다른 이는 자율주행 그 자체를 ADAS와 동급으로 구분하기도 한다. 
 양쪽의 의견도 모두 일리가 있다는 것이, ADAS와 자율주행은 감지, 중앙제어, 작동 장치로 구성되는데, 이를 사람으로 적용을 시켜보자면 감지는 눈, 중앙제어는 뇌, 작동 장치는 근육 신경계에 해당이 됩니다. 이 세 가지 요소 중 어느 것이 가장 중요하다고 말 할 수가 없는 것이, 위의 모든 구성 요소가 모두 유기적으로 작동을 해야 진정한 의미의 자율주행에 가까워지기 때문입니다.
